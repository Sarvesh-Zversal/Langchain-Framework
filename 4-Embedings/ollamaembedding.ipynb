```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding = (
    OllamaEmbeddings(model="gemma:2b")
)
```

```python
r1 = embedding.embed_documents(
    [ 
    "hello my name is sarvesh",
    "i am from up",
    ]
)
```

```python
len(r1[0]   )  # dimension of embedding
```
